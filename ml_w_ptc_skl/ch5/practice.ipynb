```python
import sklearn
data = sklearn.datasets.fetch_olivetti_faces()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
```

```python
fig, ax = plt.subplots(1,4)
for i in range(4):
    ax[i].imshow(data.images[i], cmap='Greys')
plt.show()
```

```python
data.target
```

```python
y_fac = data.target
X_fac = data.data
tsne = TSNE(n_components = 2, init='pca', random_state=123)
X_fac_tsne = tsne.fit_transform(X_fac)
```

```python
import matplotlib.patheffects as PathEffects
def plot_projection(x, targt):
    f = plt.figure(figsize=(8,8))
    ax = plt.subplot(aspect='equal')
    for i in range(39):
        plt.scatter(x[targt == i,0], x[targt==i, 1])
    for i in range(39):
        xtext, ytext = np.median(x[targt==i, :], axis=0)
        txt = ax.text(xtext, ytext, str(i), fontsize=20)
        txt.set_path_effects([PathEffects.Stroke(linewidth=5, foreground='w'), PathEffects.Normal()])
plot_projection(X_fac_tsne, y_fac)
plt.show()
```
